```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].unique())
print(total_schools)
```

```python
# Calculate the total number of students
total_students = len(school_data_complete["student_name"])
print(total_students)
```

```python
# Calculate the total budget
total_budget = sum(school_data_complete["budget"].unique())
print(total_budget)
```

```python
# Calculate the average maths score
avg_maths_score = (school_data_complete["maths_score"].mean())
print(avg_maths_score)
```

```python
# Calculate the average reading score
avg_reading_score = (school_data_complete["reading_score"].mean())
print(avg_reading_score)
```

```python
# Calculate the percentage of students with a passing maths score (50 or greater)
stu_pass_math = school_data_complete.loc[school_data_complete["maths_score"] >= 50,:]
num_stu_pass_math = len(stu_pass_math)
percent_pass_math = (num_stu_pass_math / total_students * 100)
print(percent_pass_math)
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater)
stu_pass_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 50,:]
num_stu_pass_reading = len(stu_pass_reading)
percent_pass_reading = (num_stu_pass_reading / total_students * 100)
print(percent_pass_reading)
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
stu_pass_math_reading = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50),:]
num_stu_pass_math_reading = len(stu_pass_math_reading)
percent_pass_math_reading = num_stu_pass_math_reading / total_students * 100
print(percent_pass_math_reading)
```

```python
#Create a dataframe to hold the above results
area_summary_df = pd.DataFrame({"Total Schools": [total_schools],
           "Total Students": [total_students],
          "Total Budget": [total_budget],
          "Average Maths Score": [avg_maths_score],
          "Average Reading Score": [avg_reading_score],
          "% Passing Maths": [percent_pass_math],
          "% Passing Reading": [percent_pass_reading],
          "% Overall Passing": [percent_pass_math_reading]})

area_summary_df
```

```python
# Optional: give the displayed data cleaner formatting
area_summary_df["Total Students"] = area_summary_df["Total Students"].map("{:,d}".format)
area_summary_df["Total Budget"] = area_summary_df["Total Budget"].map("${:,.2f}".format)
area_summary_df["Average Maths Score"] = area_summary_df["Average Maths Score"].map("{:,.2f}".format)
area_summary_df["Average Reading Score"] = area_summary_df["Average Reading Score"].map("{:,.2f}".format)
area_summary_df["% Passing Maths"] = area_summary_df["% Passing Maths"].map("{:,.2f}".format)
area_summary_df["% Passing Reading"] = area_summary_df["% Passing Reading"].map("{:,.2f}".format)
area_summary_df["% Overall Passing"] = area_summary_df["% Overall Passing"].map("{:,.2f}".format)

area_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths

```python
# School Name
school_name = school_data_complete.groupby("school_name")["school_name"].unique()
print(school_name)
```

```python
# School Type
school_type = school_data_complete.groupby("school_name")["type"].unique()
print(school_type)
```

```python
# Total Students
total_stu_per_school = school_data_complete.groupby("school_name")["student_name"].count()
print(total_stu_per_school)
```

```python
# Total School Budget
total_school_budget = school_data_complete.groupby("school_name")["budget"].unique()
print(total_school_budget)
```

```python
# Per Student Budget
per_stu_budget = total_school_budget / total_stu_per_school
print(per_stu_budget)
```

```python
# Average Maths Score
avg_math_score = school_data_complete.groupby("school_name")["maths_score"].mean()
print(avg_math_score)
```

```python
# Average Reading Score
avg_reading_score = school_data_complete.groupby("school_name")["reading_score"].mean()
print(avg_reading_score)
```

```python
# % Passing Maths
num_stu_pass_maths_per_school = school_data_complete[school_data_complete["maths_score"] >= 50].groupby("school_name")["Student ID"].count()

percent_stu_pass_maths_per_school = num_stu_pass_maths_per_school / total_stu_per_school * 100
print(percent_stu_pass_maths_per_school)
```

```python
# % Passing Reading
num_stu_pass_reading_per_school = school_data_complete[school_data_complete["reading_score"] >= 50].groupby("school_name")["Student ID"].count()

percent_stu_pass_reading_per_school = num_stu_pass_reading_per_school / total_stu_per_school * 100
print(percent_stu_pass_reading_per_school)
```

```python
# % Overall Passing (The percentage of students that passed maths and reading.) (ref = https://www.askpython.com/python-modules/pandas/conditionally-grouping-values)
pass_both_df = school_data_complete.query("maths_score >= 50 & reading_score >= 50")
num_stu_pass_both_per_school = pass_both_df.groupby("school_name")["Student ID"].count()

percent_stu_pass_both_per_school = num_stu_pass_both_per_school / total_stu_per_school * 100
print(percent_stu_pass_both_per_school)
```

```python
# Create a dataframe to hold the above results
per_school_summary_df = pd.DataFrame({"School type": school_type,
                         "Total Students": total_stu_per_school,
                         "Total School Budget": total_school_budget,
                         "Per Student Budget": per_stu_budget,
                         "Average Maths Score": avg_math_score,
                         "Average Reading Score": avg_reading_score,
                         "% Passing Maths": percent_stu_pass_maths_per_school,
                         "% Passing Reading": percent_stu_pass_reading_per_school,
                         "% Overall Passing": percent_stu_pass_both_per_school})
```

```python
# Format data in "School type" column (ref = https://stackoverflow.com/questions/38277928/remove-special-characters-in-pandas-dataframe)
per_school_summary_df["School type"] = per_school_summary_df["School type"].astype(str)
per_school_summary_df["School type"] = per_school_summary_df["School type"].str.replace("']", "")
per_school_summary_df["School type"] = per_school_summary_df["School type"].str.replace("['", "")

# Format data in "Total School Budget" column
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].astype(str)
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].str.replace("]", "")
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].str.replace("[", "")
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].astype(float)
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)

# Format data in "Per Student Budget" column
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].astype(str)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].str.replace("]", "")
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].str.replace("[", "")
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].astype(float)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)

per_school_summary_df
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school. 
yr9_math = school_data_complete.loc[school_data_complete["year"] == 9].groupby('school_name')["maths_score"].mean()
yr10_math = school_data_complete.loc[school_data_complete["year"] == 10].groupby('school_name')["maths_score"].mean()
yr11_math = school_data_complete.loc[school_data_complete["year"] == 11].groupby('school_name')["maths_score"].mean()
yr12_math = school_data_complete.loc[school_data_complete["year"] == 12].groupby('school_name')["maths_score"].mean()

# Create a dataframe to hold the above results
maths_scores_by_year_df = pd.DataFrame({
        "Year 9": yr9_math,
        "Year 10": yr10_math,
        "Year 11": yr11_math,
        "Year 12": yr12_math})

# Data cleaner formatting
maths_scores_by_year_df["Year 9"] = maths_scores_by_year_df["Year 9"].map("{:,.2f}".format)
maths_scores_by_year_df["Year 10"] = maths_scores_by_year_df["Year 10"].map("{:,.2f}".format)
maths_scores_by_year_df["Year 11"] = maths_scores_by_year_df["Year 11"].map("{:,.2f}".format)
maths_scores_by_year_df["Year 12"] = maths_scores_by_year_df["Year 12"].map("{:,.2f}".format)

maths_scores_by_year_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Change "Per Student Budget" data type to float
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].astype(str)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].str.replace("$", "")
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].astype(float)
per_school_summary_df.dtypes
```

```python
# Create School Spending bins
spending_bins = [0, 585, 630, 645, 680]
group_names = ["< $585", "$585-630", "$630-645", "$645-680"]
per_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary_df["Per Student Budget"], spending_bins, labels = group_names)
```

```python
# Calculate mean score per spending range
spending_math_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = per_school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = per_school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = per_school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Create a dataframe to hold the above calculation           
spending_summary_df = pd.DataFrame({
    "Average Maths Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Maths": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending})

# Data formatting
spending_summary_df["Average Maths Score"] = spending_summary_df["Average Maths Score"].map("{:,.2f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:,.2f}".format)
spending_summary_df["% Passing Maths"] = spending_summary_df["% Passing Maths"].map("{:,.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:,.2f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:,.2f}".format)

spending_summary_df
```
