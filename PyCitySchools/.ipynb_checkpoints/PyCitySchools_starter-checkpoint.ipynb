## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
## Create an overview table that summarises key metrics about each school

# School Name
school_name = school_data_complete.groupby("school_name")["school_name"].unique()
# print(school_name)

# School Type
school_type = school_data_complete.groupby("school_name")["type"].unique()
# print(school_type)

# Total Students
total_stu_per_school = school_data_complete.groupby("school_name")["student_name"].count()
# print(total_stu_per_school)

# Total School Budget
total_school_budget = school_data_complete.groupby("school_name")["budget"].unique()
# print(total_school_budget)

# Per Student Budget
per_stu_budget = total_school_budget / total_stu_per_school
# print(per_stu_budget)

# Average Maths Score
avg_math_score = school_data_complete.groupby("school_name")["maths_score"].mean()
# print(avg_math_score)

# Average Reading Score
avg_reading_score = school_data_complete.groupby("school_name")["reading_score"].mean()
# print(avg_reading_score)

# % Passing Maths (ref = https://www.statology.org/pandas-groupby-count-with-condition/)
num_stu_pass_maths_per_school = school_data_complete.groupby("school_name")["maths_score"].apply(lambda x: (x >= 50).sum())
# print(num_stu_pass_maths_per_school)
percent_stu_pass_maths_per_school = num_stu_pass_maths_per_school / total_stu * 100
# print(percent_stu_pass_maths_per_school)

# % Passing Reading (ref = https://www.statology.org/pandas-groupby-count-with-condition/)
num_stu_pass_reading_per_school = school_data_complete.groupby("school_name")["reading_score"].apply(lambda x: (x >= 50).sum())
# print(num_stu_pass_reading_per_school)
percent_stu_pass_reading_per_school = num_stu_pass_reading_per_school / total_stu * 100
# print(percent_stu_pass_reading_per_school)

# % Overall Passing (The percentage of students that passed maths and reading.) (ref = https://www.askpython.com/python-modules/pandas/conditionally-grouping-values)
pass_both_df = school_data_complete.query("maths_score >= 50 & reading_score >= 50")
num_stu_pass_both_per_school = pass_both_df.groupby("school_name")["student_name"].count()
percent_stu_pass_both_per_school = num_stu_pass_both_per_school / total_stu * 100
# print(percent_stu_pass_both_per_school)

# Create a dataframe to hold the above results
results2_df = pd.DataFrame({"School type": school_type,
                         "Total Students": total_stu_per_school,
                         "Total School Budget": total_school_budget,
                         "Per Student Budget": per_stu_budget,
                         "Average Maths Score": avg_math_score,
                         "Average Reading Score": avg_reading_score,
                         "% Passing Maths": percent_stu_pass_maths_per_school,
                         "% Passing Reading": percent_stu_pass_reading_per_school,
                         "% Overall Passing": percent_stu_pass_both_per_school})

# Format data in "School type" column (ref = https://stackoverflow.com/questions/38277928/remove-special-characters-in-pandas-dataframe)
results2_df["School type"] = results2_df["School type"].astype(str)
results2_df["School type"] = results2_df["School type"].str.replace("']", "")
results2_df["School type"] = results2_df["School type"].str.replace("['", "")

# Format data in "Total School Budget" column
results2_df["Total School Budget"] = results2_df["Total School Budget"].astype(str)
results2_df["Total School Budget"] = results2_df["Total School Budget"].str.replace("]", "")
results2_df["Total School Budget"] = results2_df["Total School Budget"].str.replace("[", "")
results2_df["Total School Budget"] = results2_df["Total School Budget"].astype(float)
results2_df["Total School Budget"] = results2_df["Total School Budget"].map("${:,.2f}".format)

# Format data in "Per Student Budget" column
results2_df["Per Student Budget"] = results2_df["Per Student Budget"].astype(str)
results2_df["Per Student Budget"] = results2_df["Per Student Budget"].str.replace("]", "")
results2_df["Per Student Budget"] = results2_df["Per Student Budget"].str.replace("[", "")
results2_df["Per Student Budget"] = results2_df["Per Student Budget"].astype(float)
results2_df["Per Student Budget"] = results2_df["Per Student Budget"].map("${:,.2f}".format)

results2_df
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school. 
yr9_math = school_data_complete.loc[school_data_complete["year"] == 9].groupby('school_name')["maths_score"].mean()
yr10_math = school_data_complete.loc[school_data_complete["year"] == 10].groupby('school_name')["maths_score"].mean()
yr11_math = school_data_complete.loc[school_data_complete["year"] == 11].groupby('school_name')["maths_score"].mean()
yr12_math = school_data_complete.loc[school_data_complete["year"] == 12].groupby('school_name')["maths_score"].mean()

avg_math_scores_per_school_df = pd.DataFrame({
        "Year 9": yr9_math,
        "Year 10": yr10_math,
        "Year 11": yr11_math,
        "Year 12": yr12_math})

# Data cleaner formatting

avg_math_scores_per_school_df["Year 9"] = avg_math_scores_per_school_df["Year 9"].map("{:,.2f}".format)
avg_math_scores_per_school_df["Year 10"] = avg_math_scores_per_school_df["Year 10"].map("{:,.2f}".format)
avg_math_scores_per_school_df["Year 11"] = avg_math_scores_per_school_df["Year 11"].map("{:,.2f}".format)
avg_math_scores_per_school_df["Year 12"] = avg_math_scores_per_school_df["Year 12"].map("{:,.2f}".format)

avg_math_scores_per_school_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create School Spending bins
bins = [0, 584.9, 629.9, 644.9, 680]
group_names = ["< $585", "$585-630", "$630-645", "$645-680"]
school_data_complete["school_spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels = group_names)

# Group by school spending bins
group_by_spending = school_data_complete.groupby("school_spending_bins")

# Calculation
avg_math = group_by_spending["maths_score"].mean()
avg_reading = group_by_spending["reading_score"].mean()
percent_pass_math = (school_data_complete[school_data_complete["maths_score"] >= 50].groupby("school_spending_bins")["Student ID"].count()/group_by_spending["Student ID"].count())*100
percent_pass_reading = (school_data_complete[school_data_complete["reading_score"] >= 50].groupby("school_spending_bins")["Student ID"].count()/group_by_spending["Student ID"].count())*100
percent_overall_pass_rate = (school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].groupby("school_spending_bins")["Student ID"].count()/group_by_spending['Student ID'].count())*100

# Create a dataframe to hold the above calculation           
scores_by_school_spending_df = pd.DataFrame({
    "Average Maths Score": avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Maths": percent_pass_math,
    "% Passing Reading": percent_pass_reading,
    "% Overall Passing": percent_overall_pass_rate})

# Data formatting

scores_by_school_spending_df["Average Maths Score"] = scores_by_school_spending_df["Average Maths Score"].map("{:,.2f}".format)
scores_by_school_spending_df["Average Reading Score"] = scores_by_school_spending_df["Average Reading Score"].map("{:,.2f}".format)
scores_by_school_spending_df["% Passing Maths"] = scores_by_school_spending_df["% Passing Maths"].map("{:,.2f}".format)
scores_by_school_spending_df["% Passing Reading"] = scores_by_school_spending_df["% Passing Reading"].map("{:,.2f}".format)
scores_by_school_spending_df["% Overall Passing"] = scores_by_school_spending_df["% Overall Passing"].map("{:,.2f}".format)

scores_by_school_spending_df
```
